Next is the api requests to get the top charting songs in rap and country

```python
NUM_PAGES = 5

chart_data_split = []

chart_data = []
chart_genres = ['rap', 'country']

for chart_genre in chart_genres:
    chart_data = []
    for i in range(1, NUM_PAGES + 1):
        successful = False
        # If a request times out, keep trying until we get it
        while (not successful):
            try:
                successful = True
                chart_data += genius.charts(time_period='all_time', chart_genre=chart_genre, per_page=50, page=i, text_format=None, type_='songs')['chart_items']
            except:
                successful = False
    chart_data_split.append(chart_data)

```

```python
# Separate out the urls and the song names from the Song object
song_urls = []
song_url_genre = []
chart_names = []

for genre in chart_data_split:
    song_url_genre = []
    chart_names_genre = []
    for i in range(len(genre)):
        song_url_genre.append(genre[i]['item']['url'])
        chart_names_genre.append(genre[i]['item']['title'])
            
    song_urls.append(song_url_genre)
    chart_names.append(chart_names_genre)
```

```python
# Check to see if there is some overlap between the songs that are manual added and the charts.
genre_num = 0
for genre in extra_songs:
    for song in genre:
        if song in chart_names[genre_num]:
            i = chart_names[genre_num].index(song)
            chart_names[genre_num].pop(i)
            song_urls[genre_num].pop(i)
            print(f"{song} already in list")
    genre_num += 1
```

```python
num_rap = len(chart_names[0]) + len(extra_songs[0])
num_country = len(chart_names[1]) + len(extra_songs[1])
print(f"{num_rap} rap songs and {num_country} country songs")
```

Finally, we will do api requests to genius to get the lyrics of each song and store these lyrics in a pandas dataframe

```python
import pandas as pd

genre = []
lyrics_list = []
 
total_songs = len(extra_songs[0]) + len(extra_songs[1]) + len(song_urls[0]) + len(song_urls[1])

songs_done = 0
for j in range(len(song_urls)):
    for i in range(len(song_urls[j])):
        successful = False
        while (not successful):
            try:
                successful = True
                lyrics= genius.lyrics( song_url=song_urls[j][i], remove_section_headers=False)
            except:
                print("Not Successful")
                successful = False
        
        lyrics_list.append(lyrics)
        genre.append(chart_genres[j])
        
        songs_done += 1
        print(f"{songs_done} out of {total_songs} done")
        
    for song in extra_songs[j]:
        successful = False
        while (not successful):
            try:
                successful = True
                lyrics = genius.search_song(song).lyrics
            except:
                print("Not Successful")
                successful = False
                
        lyrics_list.append(lyrics)
        genre.append(chart_genres[j])
        
        songs_done += 1
        print(f"{songs_done} out of {total_songs} done")
        
data = {'Lyrics': lyrics_list, 'Genre': genre}
df = pd.DataFrame.from_dict(data)
        
df.to_csv('data.csv', index=False)
```
